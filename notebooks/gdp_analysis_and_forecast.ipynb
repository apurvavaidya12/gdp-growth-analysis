```python
import pandas as pd
```

```python
df = pd.read_excel('D:/VS Code Projects/imf data/imf-dm-export-20260215.xlsx')
```

```python
df.head()
```

```python
df = df.drop(index=df.index[0])
```

```python
df.head
```

```python
df = df.rename(columns={'Real GDP growth (Annual percent change)': 'Country'})
```

```python
df.head()
```

```python
df_long = df.melt(
    id_vars='Country',
    var_name='Year',
    value_name='GDP_Growth'
)
```

```python
df_long.head()
```

```python
df_long.info()
df_long.isna().sum()
```

```python
df_long = df_long.dropna(subset=['Country',])
```

```python
df_long.info()
df_long.isna().sum()
```

```python
df_long['Year'] = df_long['Year'].astype(int)
```

```python
df_long = df_long.dropna(subset=['GDP_Growth'])
```

```python
df_long.info()
```

```python
df_long.info()
df_long.isna().sum()
```

```python
df_long.describe()
```

```python
avg_growth = (
    df_long
    .groupby('Country')['GDP_Growth']
    .mean()
    .sort_values(ascending=False)
)

avg_growth
```

```python
volatility = (
    df_long
    .groupby('Country')['GDP_Growth']
    .std()
    .sort_values(ascending=False)
)

volatility
```

```python
best_year = df_long.loc[
    df_long.groupby('Country')['GDP_Growth'].idxmax(),
    ['Country', 'Year', 'GDP_Growth']
]

best_year
```

```python
print(avg_growth)
print(volatility)
print(best_year)
```

```python
trend = (
    df_long
    .groupby('Country')
    .apply(lambda x: x['Year'].corr(x['GDP_Growth']))
    .sort_values(ascending=False)
)

trend
```

```python
summary = (
    df_long
    .groupby('Country')
    .agg(
        Avg_Growth=('GDP_Growth', 'mean'),
        Volatility=('GDP_Growth', 'std'),
        Max_Growth=('GDP_Growth', 'max'),
        Min_Growth=('GDP_Growth', 'min')
    )
    .round(2)
)

summary
```

```python
best_year = df_long.loc[
    df_long.groupby('Country')['GDP_Growth'].idxmax(),
    ['Country', 'Year', 'GDP_Growth']
].rename(columns={'GDP_Growth': 'Best_Year_Growth'})

worst_year = df_long.loc[
    df_long.groupby('Country')['GDP_Growth'].idxmin(),
    ['Country', 'Year', 'GDP_Growth']
].rename(columns={'GDP_Growth': 'Worst_Year_Growth'})

print(best_year)
print(worst_year)
```

```python
trend = (
    df_long
    .groupby('Country')
    .apply(lambda x: x['Year'].corr(x['GDP_Growth']))
    .rename('Trend_Correlation')
    .round(2)
)
trend
```

```python
final_report = (
    summary
    .merge(best_year, on='Country')
    .merge(worst_year, on='Country')
    .merge(trend, on='Country')
)

final_report
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))

for country in df_long['Country'].unique():
    data = df_long[df_long['Country'] == country]
    plt.plot(data['Year'], data['GDP_Growth'], label=country)

plt.title('GDP Growth Trend by Country')
plt.xlabel('Year')
plt.ylabel('GDP Growth (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
volatility.plot(kind='bar', figsize=(8, 5))
plt.title('GDP Growth Volatility by Country')
plt.xlabel('Country')
plt.ylabel('Standard Deviation (%)')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8, 5))

plt.scatter(final_report['Min_Growth'], final_report['Max_Growth'])

for country in final_report.index:
    plt.text(
        final_report.loc[country, 'Min_Growth'],
        final_report.loc[country, 'Max_Growth'],
        country
    )

plt.xlabel('Worst GDP Growth (%)')
plt.ylabel('Best GDP Growth (%)')
plt.title('Best vs Worst GDP Growth by Country')
plt.grid(True)
plt.show()
```

```python
import numpy as np


def forecast_country(df, country, years=10):
    data = df[df['Country'] == country]
    X = data['Year'].values
    y = data['GDP_Growth'].values

    coef = np.polyfit(X, y, 1)
    trend = np.poly1d(coef)

    future_years = np.arange(X.max() + 1, X.max() + years + 1)
    future_growth = trend(future_years)

    plt.figure(figsize=(10, 6))
    plt.plot(X, y, label='Actual')
    plt.plot(X, trend(X), linestyle='--', label='Trend')
    plt.plot(future_years, future_growth, linestyle=':', label='Forecast')
    plt.title(f'GDP Growth Forecast – {country}')
    plt.xlabel('Year')
    plt.ylabel('GDP Growth (%)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
forecast_country(df_long, 'China, People\'s Republic of')
forecast_country(df_long, 'India')
forecast_country(df_long, 'Vietnam')
forecast_country(df_long, 'Indonesia')
forecast_country(df_long, 'Malaysia')
```
